```python
# Method to get ether balance of a single account
def get_ether_balance_by_address(_address, _apiKey):
    '''
        Etherscan Single Address Balance Method
        Args::
            _address: Ethereum Address
            _apiKey: Etherscan API Key
        Returns::
            ethResJS: Etherscan ABI request result in JSON format
    '''
    ETHERSCAN_ABI_REQUEST = "https://api.etherscan.io/api?module=account&action=balance&address={0}&tag=latest&apikey={1}".format(_address, _apiKey)
    ethResJS = (requests.get(ETHERSCAN_ABI_REQUEST)).json()['result']
#     ethRes = (requests.get(ETHERSCAN_ABI_REQUEST))
    return ethResJS
```

```python
# Method to get ether balance of multiple accounts
def get_ether_balance_by_multiple_addresses(_addresses, _apiKey):
    '''
        Etherscan Multiple Address Balance Method
        Args::
            _addresses: list of Ethereum Addresses
            _apiKey: Etherscan API Key
        Returns::
            ethResJS: Etherscan ABI request result in JSON format
    '''

    # Converting to the required format : (address0,address1,address2)
    _address = ','.join(_addresses)
    ETHERSCAN_ABI_REQUEST = "https://api.etherscan.io/api?module=account&action=balancemulti&address={0}&tag=latest&apikey={1}".format(_address, _apiKey)
    ethResJS = (requests.get(ETHERSCAN_ABI_REQUEST)).json()['result']
#     ethRes = (requests.get(ETHERSCAN_ABI_REQUEST))
    return ethResJS
```

```python
# Method to fetch normal transactions of an account
def get_normal_transactions_by_address(_address, _apiKey):
    '''
        Etherscan Normal Transactions of Address Method
        Args::
            _address: Ethereum Address
            _apiKey: Etherscan API Key
        Returns::
            ethResJS: Etherscan ABI request result in JSON format
    '''
    ETHERSCAN_ABI_REQUEST = "https://api.etherscan.io/api?module=account&action=txlist&address={0}&startblock=0&endblock=99999999&page=1&offset=0&sort=asc&apikey={1}".format(_address, _apiKey)
    ethResJS = (requests.get(ETHERSCAN_ABI_REQUEST)).json()['result']
#     ethRes = (requests.get(ETHERSCAN_ABI_REQUEST))
    return ethResJS
```

```python
# Method to fetch internal transactions of an account
def get_internal_transactions_by_address(_address, _apiKey):
    '''
        Etherscan Internal Transactions of Address Method
        Args::
            _address: Ethereum Address
            _apiKey: Etherscan API Key
        Returns::
            ethResJS: Etherscan ABI request result in JSON format
    '''
    ETHERSCAN_ABI_REQUEST = "https://api.etherscan.io/api?module=account&action=txlistinternal&address={0}&startblock=0&endblock=99999999&page=1&offset=0&sort=asc&apikey={1}".format(_address, _apiKey)
    ethResJS = (requests.get(ETHERSCAN_ABI_REQUEST)).json()['result']
#     ethRes = (requests.get(ETHERSCAN_ABI_REQUEST))
    return ethResJS
```

```python
# Method to fetch token(ERC-20) transactions of an account
def get_token_transactions_by_address(_address, _apiKey):
    '''
        Etherscan ERC-20 Token Transactions of Address Method
        Args::
            _address: Ethereum Address
            _apiKey: Etherscan API Key
        Returns::
            ethResJS: Etherscan ABI request result in JSON format
    '''
    ETHERSCAN_ABI_REQUEST = "https://api.etherscan.io/api?module=account&action=tokentx&address={0}&startblock=0&endblock=99999999&page=1&offset=0&sort=asc&apikey={1}".format(_address, _apiKey)
    ethResJS = (requests.get(ETHERSCAN_ABI_REQUEST)).json()['result']
#     ethRes = (requests.get(ETHERSCAN_ABI_REQUEST))
    return ethResJS
```

```python
# Method to fetch NFT(ERC-721) transactions of an account
def get_nft_transactions_by_address(_address, _apiKey):
    '''
        Etherscan ERC-721 Token Transactions of Address Method
        Args::
            _address: Ethereum Address
            _apiKey: Etherscan API Key
        Returns::
            ethResJS: Etherscan ABI request result in JSON format
    '''
    ETHERSCAN_ABI_REQUEST = "https://api.etherscan.io/api?module=account&action=tokennfttx&address={0}&startblock=0&endblock=99999999&page=1&offset=0&sort=asc&apikey={1}".format(_address, _apiKey)
    ethResJS = (requests.get(ETHERSCAN_ABI_REQUEST)).json()['result']
#     ethRes = (requests.get(ETHERSCAN_ABI_REQUEST))
    return ethResJS
```

```python
# Looping through all the wallet addresses and storing the pickle files

for account in wallet_addresses:
    # Fetching Normal Transactions
    normal_transactions = get_normal_transactions_by_address(account, ETHERSCAN_API_KEY)
    # Fetching Normal Transactions
    internal_transactions = get_internal_transactions_by_address(account, ETHERSCAN_API_KEY)
    # Fetching Normal Transactions
    token_transactions = get_token_transactions_by_address(account, ETHERSCAN_API_KEY)
    # Fetching Normal Transactions
    nft_transactions = get_nft_transactions_by_address(account, ETHERSCAN_API_KEY)

    # Filename for various transactions categories
    normal_transactions_filename = './wallet_transactions/normal_transactions/' + str(account) + '.csv'
    internal_transactions_filename = './wallet_transactions/internal_transactions/' + str(account) + '.csv'
    token_transactions_filename = './wallet_transactions/token_transactions/' + str(account)  + '.csv'
    nft_transactions_filename = './wallet_transactions/nft_transactions/' + str(account) + '.csv'

    # Saving normal transactions
    df_normal = pd.DataFrame(normal_transactions)
    df_normal.to_csv(normal_transactions_filename)

    # Saving internal transactions
    df_internal = pd.DataFrame(internal_transactions)
    df_internal.to_csv(internal_transactions_filename)
    
    # Saving token transactions
    df_token = pd.DataFrame(token_transactions)
    df_token.to_csv(token_transactions_filename)
    
    # Saving NFT transactions
    df_nft = pd.DataFrame(nft_transactions)
    df_nft.to_csv(nft_transactions_filename)
```
